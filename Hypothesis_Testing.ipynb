#### <b>1 .</b> A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.

```python
import pandas as pd
cutlets=pd.read_csv('Cutlets.csv')
cutlets.head()
```

**To analyze the data and draw inferences, we can use a two-sample t-test**

**Null Hypothesis (H0):  μ1=μ2**

**There is no significant difference in the diameter of the cutlets between the two units.**
<br>
<br>
**Alternative Hypothesis (H1): μ1 is not equal to μ2**  

**There is a significant difference in the diameter of the cutlets between the two units.**

```python
from scipy import stats
alpha=0.05 #sigficance level
stats.ttest_ind(cutlets['Unit A'],cutlets['Unit B'])

```

**Here p_value is greater than alpha so we fail to reject null hypothesis**

**There is no significant difference in the diameter of the cutlets between the two units.**

#### <b>2 .</b> A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.

#### Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

**we can perform an Analysis of Variance (ANOVA) test. The ANOVA test is used to compare the means of more than two groups**

**H0:μ1=μ2=μ3=μ4**

**Ha:At least one μi is different  μ**

```python
tat=pd.read_csv('LabTAT.csv')
tat.head()
```

```python
stats.f_oneway(tat['Laboratory 1'],tat['Laboratory 2'],tat['Laboratory 3'],tat['Laboratory 4'])
```

**p-value (2.11 * 10^-57 which is almost equal to zero) less than the chosen significance level (0.05), we reject the null hypothesis and conclude that there is a significant difference in average TAT among the laboratories**

#### <b>4 . </b>Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
br=pd.read_csv('BuyerRatio.csv')
br
```

**H0 : The male-female buyer rations are similar across regions**


**H1 : The male-female buyer rations are not similar across regions**

```python
br.set_index('Observed Values',inplace=True)
br.values
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(br.values)
```

**P value of Chi-Square test is 0.66 > alpha**

**Fail to reject null hypothesis**

**The male-female buyer rations are similar across regions**

#### <b>4 . </b>TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain %  of the customer order forms. Any error in order form renders it defective and has to be reworked before processing.  The manager wants to check whether the defective %  varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences

```python
cof=pd.read_csv('Costomer+OrderForm.csv')
cof.head()
```

**H0 : The defective % does not varies by centre**

**H1 : The defective % does varies by centre**

```python
cof.Phillippines.value_counts()
```

```python
dd=cof.melt(var_name='Center', value_name='Condition')
dd
```

```python
contingency_table=pd.crosstab(
    dd['Center'],
    dd['Condition']
)
```

```python
alpha=0.05
stats.chi2_contingency(contingency_table)
```

**P value of Chi-Square test is 0.277 > alpha**

**Fail to reject null hypothesis**

**There is no association between the center and the defective status of order forms**
